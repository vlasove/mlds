```python
import pandas as pd 
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('close_prices.csv', index_col='date')
df.head()
```

```python
"""
PCA fitting
"""
pca = PCA(n_components=10)
pca.fit(df)
```

```python
"""
PCA search
"""
total = 0
boarder = 0.9
counter = 0
for ratio in pca.explained_variance_ratio_:
    if total >= boarder:
        break
    counter += 1
    total += ratio

with open('w4s1.txt', 'w') as fhand:
    fhand.write(str(counter))
```

```python
"""
Pca using
"""
import numpy as np
df_pca = pca.transform(df)
first_component = pd.DataFrame(df_pca[:, 0])

df_dj = pd.read_csv('index.csv', index_col='date')
coef = np.corrcoef(first_component.T, df_dj.T)[1, 0]

with open('w4s2.txt', 'w') as fhand:
    fhand.write(str(np.round(coef, decimals=2)))
```
